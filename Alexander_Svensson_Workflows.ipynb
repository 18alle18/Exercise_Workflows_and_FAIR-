```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read data
df = pd.read_csv(r'simulated_bird_counts.csv')

# Extract variables
x = df['yr']
y = df['count']

# Plot
plt.plot(x[0:7],y[0:7], 'o')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Bird count per year')
plt.grid()
plt.show()
```
